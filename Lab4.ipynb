```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

#create a synthetic dataset
np.random.seed(42)
data= np.random.normal(0,1,100)
outliers = np.random.uniform(5,10,5)
```

```python
data_withoutliers = np.concatenate((data,outliers))
df = pd.DataFrame(data_withoutliers, columns=['values'])

Q1= df['values'].quantile(0.25)
Q3= df['values'].quantile(0.75)
IQR = Q3-Q1
lower_bound= Q1-1.5*IQR
upper_bound= Q3+1.5*IQR

outliers_IQR = df[(df['values']< lower_bound)| (df['values']> upper_bound)]
```

```python
outliers_IQR
```

```python
from matplotlib import pyplot as plt
outliers_IQR['values'].plot(kind='line', figsize=(8, 4), title='values')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
from matplotlib import pyplot as plt
outliers_IQR['values'].plot(kind='hist', bins=20, title='values')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
